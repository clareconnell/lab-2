### Visibility of system status
The system should always keep users informed about what is going on, through appropriate feedback within reasonable time.

  The American Eagle website is very successful in adhering to visibility of system status. The website gives great feedback when a mouse is placed over any icon. For example, when the user places their mouse over the favorites button on the homepage, the oulined heart becomes a solid black heart. This notifies the user that the system has understood thier action and notifies the user that if they click on it, they will be tranported to their favorites list. The first selections that the user can choose from on the homepage can be understood by any viewer. When the user clicks on an item they would like to purcahse, when the user scrolls down they can view the store's "size chart." Users often choose system functions by mistake and will need a clearly marked "emergency exit" to leave the unwanted state without having to go through an extended dialogue. Support undo and redo.

A place within the American Eagle website that needs improvements is when the user gets to their shopping bag. When the user adds an item to their bag and chooses to remove it, once the action is complete they cannot undo thier decision. While the website did make the button to remove an item from the bag small to try and prevent errors, there is no "emergecy exit" established to help the user. ### Consistency and standards
Users should not have to wonder whether different words, situations, or actions mean the same thing. Follow platform conventions.

![bag](https://raw.githubusercontent.com/clareconnell/lab-2/master/images/Screen%20Shot%202020-09-17%20at%204.55.54%20PM.png)
  The webpage uses consistent words, symbols, and labels that the user is familiar with in other webpages they have been to. There are no phrases this webpage uses that would confuse the user. They also have placed icons such as the shopping bag in the upper right hand corner, which is standard for the placement of a shopping bag within all websites. This allows the user to easily explore the webpage without become confused by where something is or by what something means. 

### Error prevention
Even better than good error messages is a careful design which prevents a problem from occurring in the first place. Either eliminate error-prone conditions or check for them and present users with a confirmation option before they commit to the action.

![fullbag](https://raw.githubusercontent.com/clareconnell/lab-2/master/images/Screen%20Shot%202020-09-17%20at%204.51.17%20PM.png) ![emptybag](https://raw.githubusercontent.com/clareconnell/lab-2/master/images/Screen%20Shot%202020-09-17%20at%204.52.34%20PM.png)

A place within the website that does not do a great job adhering to error prevention is the shopping bag. When a user removes an item from their bag, it instantly removes the item. If a consumer is clicking quickly and misclicks the remove button, there is no notification that pops up to prevent a problem. The website should create a pop-up that would ask the user if they are sure they want to commit to thier action. This would help eliminate frusteration for the user and save them time.

### Recognition rather than recall
Error messages should be expressed in plain language (no codes), precisely indicate the problem, and constructively suggest a solution.

![list](https://raw.githubusercontent.com/clareconnell/lab-2/master/images/Screen%20Shot%202020-09-17%20at%205.02.35%20PM.png)

A place within the website that contains a great example of how the website implemented recognition rather than recall is the different break down of products that the user can chose from when looking at the men or women's tab. They have keywords such as tops, jeans, loungewear, jackets, shoes, etc. This allows the user to recognize a category they would like to view without having to recall and search a category from their memory. Also, another example is if one clicks on their shopping bag and it is empty, the webpage offers three different categories of clothing they could look at instead of them having to think about what category they might want to explore next.
